```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd. read_csv('spotify-2023.csv', encoding='cp1251')

#1 побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 найпопулярніших пісень на Spotify у 2023 році за кількістю прослуховувань. 
songs_2023 = data[data["released_year"] == 2023]

top_songs = songs_2023.sort_values(by="streams", ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_songs["track_name"], top_songs["streams"], color="red")
plt.xlabel("Кількість прослуховувань")
plt.ylabel("Пісня")
plt.title("10 найпопулярніших пісень на Spotify у 2023 році")
plt.gca().invert_yaxis()
plt.show
```

```python
#2 cтворіть кругову діаграму або секторну діаграму, яка відображає місяць випуску  найпопулярніших пісень у 2023 році

grouped_data = songs_2023.groupby("released_month").size()


months = [
    "Січень", "Лютий", "Березень", "Квітень", "Травень", "Червень",
    "Липень", "Серпень", "Вересень", "Жовтень", "Листопад", "Грудень"
]


plt.figure(figsize=(8, 8))
plt.pie(grouped_data, autopct='%1.1f%%', startangle=140, pctdistance=0.85)


plt.legend(months, title="Місяці", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title("Розподіл пісень за місяцем випуску у 2023 році")

plt.show()
```

```python
#3 побудуйте лінійний графік, який відображає зміни кількості прослуховувань певного треку протягом року 2023 року.


selected_columns = ["track_name", "released_month", "streams"]
songs_2023_selected = songs_2023[selected_columns]


grouped_data = songs_2023_selected.groupby("track_name")


plt.figure(figsize=(12, 6))

for track_name, track_data in grouped_data:
    plt.plot(
        track_data['released_month'],
        track_data['streams'],
        marker='o',
        linestyle='-',
        label=track_name
    )

plt.xlabel("Місяць випуску")
plt.ylabel("Кількість прослуховувань")
plt.title("Динаміка популярності треків у 2023 році")
plt.xticks(songs_2023_selected["released_month"].unique())
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1), fontsize='small')
plt.grid(True)
plt.show()

#не дуже вийшло виконати це завдання,але намагалась


```

```python
#4 створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших артистів на Spotify у 2023 році за кількістю слухачів.
data = data.sort_values(by='streams', ascending=False)


top_10_artists = data.head(10)


plt.figure(figsize=(12, 6))
plt.barh(top_10_artists['artist(s)_name'], top_10_artists['streams'], color="blue")
plt.xlabel("Кількість прослуховувань")
plt.ylabel("Артисти")
plt.title("10 найпопулярніших артистів на Spotify у 2023 році")
plt.gca().invert_yaxis()
plt.show()
```

```python
#5  побудуйте горизонтальну стовпчикову діаграму, яка порівнює кількість прослуховувань трьох конкретних треків у 2023 році.
tracks = data.sample(n=3)
plt.figure(figsize=(10, 4))
plt.barh(tracks["track_name"], tracks["in_spotify_playlists"], color="grey")
plt.xlabel("Кількість прослуховувань")
plt.ylabel("Назва треку")
plt.title("Порівняння популярності трьох треків на Spotify у 2023 році")
plt.gca().invert_yaxis()
plt.show()
```
