```python
# 1   (scatter plot)    (iris dataset)

from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]
print(iris_df.head())

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal width (cm)', y='sepal length (cm)', size='petal length (cm)',
                hue='species', palette='Set2', data=iris_df, sizes=(10, 100), legend='full')
plt.xlabel('  (cm)')
plt.ylabel('  (cm)')
plt.title(' ')

plt.show()
```

```python
# 2 C   (scatter plot)     
data = {
    'nation': ['USA', 'Germany', 'China', 'USA', 'Germany', 'China'],
    'medal': ['Gold', 'Gold', 'Gold', 'Silver', 'Silver', 'Silver'],
    'count': [46, 27, 26, 37, 23, 18]
}
medals_df = pd.DataFrame(data)
marker_size = 50
plt.figure(figsize=(6, 2))
sns.scatterplot(x='count', y='nation', hue='medal', style='medal',
                palette='Set1', s=marker_size, data=medals_df, legend='full')
plt.xlabel('  ')
plt.ylabel('')
plt.title('        ')
plt.show()
```

```python
# 3   (scatter plot)        
data = {
    'total_bill': [16.99, 10.34, 21.01, 23.68, 24.59],
    'tip': [1.01, 1.66, 3.5, 3.31, 3.61]
}

tips_df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=tips_df, scatter_kws={"s": 100})
plt.xlabel('  ')
plt.ylabel(' ')
plt.title('\'    ')
plt.show()
```

```python
# 4    (bar chart)        "" (year)  "" (population)
years = [2000, 2005, 2010, 2015, 2020]
population = [30412500, 32051800, 34126000, 36541000, 37742154]
plt.figure(figsize=(8, 6))
plt.bar(years, population, color='skyblue')
plt.xlabel('')
plt.ylabel(' ')
plt.title('    ')
plt.show()
```

```python
# 5    (pie chart)       "tip"   "day" ()
tips = [20, 15, 25, 18]
days = ['', '', '', '']
plt.figure(figsize=(8, 8))
plt.pie(tips, labels=days, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'yellowgreen', 'coral', 'gold'])
plt.title('   ')
plt.show()
```
