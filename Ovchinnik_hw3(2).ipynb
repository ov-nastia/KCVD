```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd


df = pd.read_csv('shopping_trends.csv')


print("Розмір таблиці:", df.shape)


print("Наявність пропусків:")
print(df.isnull().sum())


print("Типи стовпців:")
print(df.dtypes)


df.columns = df.columns.str.lower().str.replace(' ', '_')


print("Оновлені назви стовпців:")
print(df.columns)

```

```python
#1 Побудуйте гістограму віку клієнтів, використовуючи Seaborn. 

sns.histplot(df['age'], bins=20, kde=True, color='purple')  

plt.title('Гістограма віку клієнтів')
plt.xlabel('Вік')
plt.ylabel('Кількість клієнтів')
plt.show()
```

```python
#2 Створіть стовпчасту діаграму або графік розподілу, щоб порівняти, які товари покупають клієнти найчастіше. 

items_purchased = df['item_purchased'].value_counts()

plt.figure(figsize=(12, 6))


items_purchased.plot(kind='bar', color='black')


plt.title('Розподіл покупок за статтею')
plt.xlabel('Стаття')
plt.ylabel('Кількість покупок')
plt.tight_layout()
plt.show()
```

```python
#3 Використовуйте scatter plot, щоб візуалізувати залежність між віком та сумою покупок. 

plt.figure(figsize=(10, 6))

plt.scatter(df['age'], df['purchase_amount_(usd)'], c='Magenta', alpha=0.5)  


plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупок')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4 Побудуйте графік розподілу оцінок товарів для оцінки якості. 
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['review_rating'], color='Salmon')  
plt.title('Розподіл оцінок товарів')
plt.xlabel('Оцінка')
plt.show
```

```python
#5 Створіть категоріальну діаграму, щоб відобразити розподіл покупок за статтею. 

plt.figure(figsize=(8, 6))

sns.countplot(data=df, x='gender', color="Lime") 

plt.title('Розподіл покупок за статтею')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
#6 Вивчіть, чи існує зв'язок між рейтингом товару і сумою покупок за допомогою scatter plot.
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="review_rating", y="purchase_amount_(usd)", alpha=0.6, color="Crimson")
plt.title("Залежність між рейтингом товару та сумою покупок")
plt.xlabel("Рейтинг товару")
plt.ylabel("Сума покупок")
plt.show()
```

```python
#7 Побудуйте стовпчату діаграму, яка відображає, в який сезон клієнти роблять найбільше покупок.
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="season", color="grey")
plt.title("Сезонність покупок")
plt.xlabel("Сезон")
plt.ylabel("Кількість покупок")
plt.show()
```

```python
#8 Створіть графік розподілу розмірів одягу, щоб визначити популярність різних розмірів. 
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x="size", color="Cyan")
plt.title("Розподіл розмірів одягу")
plt.xlabel("Розмір одягу")
plt.ylabel("Кількість покупок")
plt.show()
```

```python
#9 Порівняйте, скільки покупок робиться за різними методами оплати за допомогою стовпчатої діаграми. 
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="payment_method", color="Teal")
plt.title("Кількість покупок відносно типу оплати")
plt.xlabel("Тип оплати")
plt.ylabel("Кількість покупок")
plt.xticks(rotation=45)  
plt.show()
```

```python
#10 Створіть scatter plot для визначення, чи існує зв'язок між кількістю попередніх покупок та сумою поточної покупки.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="previous_purchases", y="purchase_amount_(usd)", alpha=0.6, color="Gold")
plt.title("Залежність між покупками")
plt.xlabel("Кількість попередніх покупок")
plt.ylabel("Сума поточної покупки")
plt.show()
```

Побудуйте графік поверхні (3D surface plot) для візуалізації залежності між трьома числовими змінними з використанням бібліотеки Seaborn. 

Налаштуйте відображення графіка, включаючи назву графіку, підписи осей і інші необхідні параметри для кращого розуміння візуалізації. 

Додайте кольорову легенду. 


```python
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")
surface = ax.plot_surface(x, y, z, cmap="plasma")
ax.set_title("3D Surface Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10) 
plt.show()
```
